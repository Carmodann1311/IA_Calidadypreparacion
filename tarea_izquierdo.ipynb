```python
import numpy as np
import matplotlib.pyplot as plt

def simular_variable_aleatoria_normal(media, desviacion_estandar, n):
    # Generar n números aleatorios con distribución normal
    valores_aleatorios = np.random.normal(media, desviacion_estandar, n)
    return valores_aleatorios

# Parámetros de las distribuciones normales
medias = [8.49, 34.64, 54.85]  # Medias de las distribuciones
desviaciones_estandar = [1.34, 11.74, 10.54]  # Desviaciones estándar de las distribuciones
n = 1000  # Número de muestras a simular para cada distribución

# Simular las tres variables aleatorias normales
variables_aleatorias = [simular_variable_aleatoria_normal(media, desviacion_estandar, n)
                        for media, desviacion_estandar in zip(medias, desviaciones_estandar)]

# Graficar las tres variables aleatorias juntas
plt.figure(figsize=(10, 6))
for i in range(3):
    plt.hist(variables_aleatorias[i], bins=30, density=True, alpha=0.6, label=f'Variable {i+1}')

plt.title('Histograma de tres variables aleatorias normales')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

def simular_variable_aleatoria_normal(media, desviacion_estandar, n):
    # Generar n números aleatorios con distribución normal
    valores_aleatorios = np.random.normal(media, desviacion_estandar, n)
    return valores_aleatorios

# Parámetros de las distribuciones normales
medias = [9.66, 33.07, 53.76]  # Medias de las distribuciones
desviaciones_estandar = [1.38, 11.52, 10.99]  # Desviaciones estándar de las distribuciones
n = 1000  # Número de muestras a simular para cada distribución

# Simular las tres variables aleatorias normales
variables_aleatorias = [simular_variable_aleatoria_normal(media, desviacion_estandar, n)
                        for media, desviacion_estandar in zip(medias, desviaciones_estandar)]

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame({'Variable 1': variables_aleatorias[0],
                   'Variable 2': variables_aleatorias[1],
                   'Variable 3': variables_aleatorias[2]})

# Calcular estadísticas resumidas
resumen_estadistico = df.describe()

# Mostrar la tabla dinámica
print("Tabla dinámica de las variables aleatorias:")
print(resumen_estadistico)
```
