```python
import numpy as np
import matplotlib.pyplot as plt

def simular_variable_aleatoria_normal(media, desviacion_estandar, n):
    # Generar n números aleatorios con distribución normal
    valores_aleatorios = np.random.normal(media, desviacion_estandar, n)
    return valores_aleatorios

# Parámetros de las distribuciones normales
medias = [8.33, 34.14, 54.86]  # Medias de las distribuciones
desviaciones_estandar = [1.18, 11.95, 11.37]  # Desviaciones estándar de las distribuciones
n = 1000  # Número de muestras a simular para cada distribución

# Simular las tres variables aleatorias normales
variables_aleatorias = [simular_variable_aleatoria_normal(media, desviacion_estandar, n)
                        for media, desviacion_estandar in zip(medias, desviaciones_estandar)]

# Graficar las tres variables aleatorias juntas
plt.figure(figsize=(10, 6))
for i in range(3):
    plt.hist(variables_aleatorias[i], bins=30, density=True, alpha=0.6, label=f'Variable {i+1}')

plt.title('Histograma de tres variables aleatorias normales')
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

def simular_variable_aleatoria_normal(media, desviacion_estandar, n):
    # Generar n números aleatorios con distribución normal
    valores_aleatorios = np.random.normal(media, desviacion_estandar, n)
    return valores_aleatorios

# Parámetros de las distribuciones normales
medias = [9.66, 33.07, 53.76]  # Medias de las distribuciones
desviaciones_estandar = [1.38, 11.52, 10.99]  # Desviaciones estándar de las distribuciones
n = 1000  # Número de muestras a simular para cada distribución

# Simular las tres variables aleatorias normales
variables_aleatorias = [simular_variable_aleatoria_normal(media, desviacion_estandar, n)
                        for media, desviacion_estandar in zip(medias, desviaciones_estandar)]

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame({'Variable 1': variables_aleatorias[0],
                   'Variable 2': variables_aleatorias[1],
                   'Variable 3': variables_aleatorias[2]})

# Calcular estadísticas resumidas
resumen_estadistico = df.describe()

# Mostrar la tabla dinámica
print("Tabla dinámica de las variables aleatorias:")
print(resumen_estadistico)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Medias y desviaciones estándar proporcionadas
medias = [8.33, 34.14, 54.86]
desviaciones_estandar = [1.18, 11.95, 11.37]

# Número de muestras por distribución
n_samples = 1000

# Generar las muestras de datos
data1 = np.random.normal(loc=medias[0], scale=desviaciones_estandar[0], size=n_samples)
data2 = np.random.normal(loc=medias[1], scale=desviaciones_estandar[1], size=n_samples)
data3 = np.random.normal(loc=medias[2], scale=desviaciones_estandar[2], size=n_samples)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, edgecolor='black', alpha=0.5, label=f'Media: {medias[0]}, Desviación: {desviaciones_estandar[0]}')
plt.hist(data2, bins=30, edgecolor='black', alpha=0.5, label=f'Media: {medias[1]}, Desviación: {desviaciones_estandar[1]}')
plt.hist(data3, bins=30, edgecolor='black', alpha=0.5, label=f'Media: {medias[2]}, Desviación: {desviaciones_estandar[2]}')

# Añadir leyenda y etiquetas
plt.title('Histograma de las tres distribuciones')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
