```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos proporcionados
data = {
    'Club': ["Manchester City", "Paris Saint-Germain", "Manchester United", "Chelsea FC", "Liverpool FC", "Clube Atlético Mineiro", "FC Augsburg", "Lokomotiv Moscow", "Trabzonspor", "CA Osasuna"],
    'Market_value': ["€1.05bn", "€997.75m", "€937.25m", "€882.50m", "€879.50m", "€98.45m", "€97.10m", "€97.03m", "€96.30m", "€93.50m"]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Limpiar los valores de Market_value y convertirlos a números
df['Market_value'] = df['Market_value'].replace({'€': '', 'bn': 'e9', 'm': 'e6'}, regex=True).astype(float)

# Graficar
plt.figure(figsize=(10, 6))
plt.bar(df['Club'], df['Market_value'], color='skyblue')
plt.xlabel('Club')
plt.ylabel('Market Value (€)')
plt.title('Market Value de los Clubs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
