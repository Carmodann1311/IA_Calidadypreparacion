```python
import pandas as pd

# URL de la tabla general
url = 'https://www.mediotiempo.com/futbol/liga-mx-femenil/tabla-general'

# Lee la tabla en un DataFrame
df = pd.read_html(url)[0]

# Muestra la tabla
print(df)
```

```python
# Renombra las columnas
df.rename(columns={'POS': 'Posición', 'JJ': 'Juegos jugados', 'JG': 'Juegos ganados',
                   'JE': 'Juegos empatados', 'JP': 'Juegos perdidos',
                                      'GF': 'Goles a favor', 'GC': 'Goles en contra', 'PTS': 'Puntos'}, inplace=True)
```

```python
# Guarda en un archivo CSV
df.to_csv('liga_mx_femenil_clausura_2024.csv', index=False)
```
