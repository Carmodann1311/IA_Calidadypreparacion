***Paso 3: Gr치fica del Proceso Poisson***

```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metros del proceso de Poisson
lambda_poisson = 2  # tasa de ocurrencia promedio por unidad de tiempo
tiempo_total = 10  # tiempo total de observaci칩n
intervalos = 1000  # n칰mero de intervalos de tiempo

# Generaci칩n del proceso de Poisson
tiempos = np.linspace(0, tiempo_total, intervalos)
eventos = np.random.poisson(lambda_poisson * (tiempos[1] - tiempos[0]), size=intervalos)
conteo_acumulado = np.cumsum(eventos)

# Gr치fica del proceso de Poisson
plt.figure(figsize=(10, 6))
plt.step(tiempos, conteo_acumulado, where='post')
plt.xlabel('Tiempo')
plt.ylabel('N칰mero acumulado de eventos')
plt.title('Proceso de Poisson')
plt.grid(True)
plt.show()
```

***Paso 4: C치lculo de las probabilidades***
Probabilidades
Las probabilidades de que el n칰mero de eventos Xt en el tiempo total de observaci칩n (10 unidades de tiempo) sea igual a 1, 2, 3, 4 o 5 son las siguientes:


```python
from scipy.stats import poisson

# Par치metros
lambda_t = lambda_poisson * tiempo_total  # lambda * t

# C치lculo de las probabilidades
probabilidades = [poisson.pmf(i, lambda_t) for i in range(1, 6)]

# Mostrar resultados
probabilidades
```
