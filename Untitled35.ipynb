```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
lambda_rate = 5  # tasa de llegadas por hora
t = 1  # intervalo de tiempo en horas
```

```python
np.random.seed(42)  # Para reproducibilidad
event_times = np.random.poisson(lambda_rate, size=1000)
```

```python
time_points = np.arange(0, 1000)
cumulative_events = np.cumsum(event_times)

plt.figure(figsize=(10, 6))
plt.step(time_points, cumulative_events, where='post')
plt.xlabel('Tiempo (unidades arbitrarias)')
plt.ylabel('Número acumulado de eventos')
plt.title('Proceso Poisson con λ = 5')
plt.grid(True)
plt.show()
```

```python
probabilities = {i: poisson.pmf(i, lambda_rate * t) for i in range(1, 6)}
print(probabilities)
```
