```python
promedio_edad = df['Age'].mean()
edad_jugador_joven = df['Age'].min()
edad_jugador_veterano = df['Age'].max()
promedio_salario = df['Salary'].mean()
salario_mas_bajo = df['Salary'].min()
salario_mas_alto = df['Salary'].max()
print(f"Promedio de edad de los jugadores: {promedio_edad:.2f}")
print(f"Edad del jugador más joven: {edad_jugador_joven}")
print(f"Edad del jugador más veterano: {edad_jugador_veterano}")
print(f"Promedio de salario de los jugadores: {promedio_salario:.2f}")
print(f"Salario más bajo de los jugadores: {salario_mas_bajo}")
print(f"Salario más alto de los jugadores: {salario_mas_alto}")
```

```python
moda_numero = df['Number'].mode()
print(f"La moda de los números que utilizan los jugadores es: {moda_numero.values}")
```

```python
jugadores_por_equipo = df['Team'].value_counts()
equipo_con_menos_jugadores = jugadores_por_equipo.idxmin()
equipo_con_mas_jugadores = jugadores_por_equipo.idxmax()
print(f"Equipo con menos jugadores: {equipo_con_menos_jugadores}")
print(f"Equipo con más jugadores: {equipo_con_mas_jugadores}")
```

```python
frecuencia_college = df['College'].value_counts()
frecuencia_college_ordenada = frecuencia_college.sort_values(ascending=False)
print("Frecuencia de las variables categóricas en la columna 'College', ordenadas de mayor a menor:")
print(frecuencia_college_ordenada)

```

```python
frecuencia_college = df['College'].value_counts()
universidad_menos_repetida = frecuencia_college.idxmin()
cantidad_menos_repetida = frecuencia_college.min()
universidad_mas_repetida = frecuencia_college.idxmax()
cantidad_mas_repetida = frecuencia_college.max()
print(f"Universidad que se repite menos: {universidad_menos_repetida} con {cantidad_menos_repetida} estudiantes")
print(f"Universidad que se repite más: {universidad_mas_repetida} con {cantidad_mas_repetida} estudiantes")
```

```python
numero_mas_repetido = df['Number'].value_counts().idxmax()
moda_numero_mas_usado = df[df['Number'] == numero_mas_repetido]['Number'].mode().iloc[0]
print(f"Most used number: {numero_mas_repetido}")
print(f"Least used number: {numero_menos_usado}")
print(f"Mode of the most used number: {moda_numero_mas_usado}")
```

```python
if "Team" in df.columns:
    promedio_estatura_por_equipo = df.groupby('Team')['Height'].mean()
    print("Promedio de estatura por equipos:")
    print(promedio_estatura_por_equipo)
else:
    print("Column 'Team' not found in the DataFrame.")
```

```python
promedio_estatura_por_posicion = df.groupby('Position')['Height'].mean()
posicion_promedio_estatura_mas_baja = promedio_estatura_por_posicion.idxmin()
promedio_estatura_mas_baja = promedio_estatura_por_posicion.min()
posicion_promedio_estatura_mas_alta = promedio_estatura_por_posicion.idxmax()
promedio_estatura_mas_alta = promedio_estatura_por_posicion.max()

print(f"La posición con el promedio de estatura más baja es '{posicion_promedio_estatura_mas_baja}' con un promedio de {promedio_estatura_mas_baja} pulgadas.")
print(f"La posición con el promedio de estatura más alta es '{posicion_promedio_estatura_mas_alta}' con un promedio de {promedio_estatura_mas_alta} pulgadas.")
```

```python
estatura_mas_baja_por_posicion = df.groupby('Position')['Height'].min()
estatura_mas_alta_por_posicion = df.groupby('Position')['Height'].max()

print("Estatura más baja por posición:")
print(estatura_mas_baja_por_posicion)
print("\nEstatura más alta por posición:")
print(estatura_mas_alta_por_posicion)
```

```python
promedio_salario_por_posicion = df.groupby('Position')['Salary'].mean()
posicion_salario_mas_bajo = promedio_salario_por_posicion.idxmin()
promedio_salario_mas_bajo = promedio_salario_por_posicion.min()

posicion_salario_mas_alto = promedio_salario_por_posicion.idxmax()
promedio_salario_mas_alto = promedio_salario_por_posicion.max()

print(f"La posición con el salario más bajo en promedio es '{posicion_salario_mas_bajo}' con un promedio de ${promedio_salario_mas_bajo:.2f}.")
print(f"La posición con el salario más alto en promedio es '{posicion_salario_mas_alto}' con un promedio de ${promedio_salario_mas_alto:.2f}.")
```
